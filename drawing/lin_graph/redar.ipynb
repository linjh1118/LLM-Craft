```python
# Libraries
import matplotlib.pyplot as plt
import pandas as pd
from math import pi
 
# Set data
df = pd.DataFrame({
'group': ['A','B','C','D'],
'var1': [38, 1.5, 30, 4],
'var2': [29, 10, 9, 34],
'var3': [8, 39, 23, 24],
'var4': [7, 31, 33, 14],
'var5': [28, 15, 32, 14]
})
 
# ------- PART 1: Create background
 
# number of variable
categories=list(df)[1:]
N = len(categories)
 
# What will be the angle of each axis in the plot? (we divide the plot / number of variable)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
 
# Initialise the spider plot
ax = plt.subplot(111, polar=True)
 
# If you want the first axis to be on top:
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)
 
# Draw one axe per variable + add labels
plt.xticks(angles[:-1], categories)
 
# Draw ylabels
ax.set_rlabel_position(0)
plt.yticks([10,20,30], ["10","20","30"], color="grey", size=7)
plt.ylim(0,40)
 

# ------- PART 2: Add plots
 
# Plot each individual = each line of the data
# I don't make a loop, because plotting more than 3 groups makes the chart unreadable
 
# Ind1
values=df.loc[0].drop('group').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="group A")
ax.fill(angles, values, 'b', alpha=0.1)
 
# Ind2
values=df.loc[1].drop('group').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="group B")
ax.fill(angles, values, 'r', alpha=0.1)
 
# Add legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# Show the graph
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 假设的评分数据，您可以用实际的数据替换这些值
categories = ['条理性', '趣味性', '性格一致性']
model_A_scores = {
    'Good': [70, 55, 65],
    'Same': [20, 30, 25],
    'Bad': [10, 15, 10]
}
model_B_scores = {
    'Good': [30, 45, 35],
    'Same': [20, 30, 25],
    'Bad': [50, 25, 40]
}

# 准备数据
labels = categories
A_good = model_A_scores['Good']
A_same = model_A_scores['Same']
A_bad = model_A_scores['Bad']
B_good = model_B_scores['Good']
B_same = model_B_scores['Same']
B_bad = model_B_scores['Bad']

x = np.arange(len(labels))  # 标签位置
width = 0.35  # 条形图宽度

fig, ax = plt.subplots(figsize=(10, 6))

# 绘制模型A的条形图
rects1 = ax.bar(x - width/2, A_good, width, label='Model A Good', color='b')
rects2 = ax.bar(x - width/2, A_same, width, bottom=A_good, label='Model A Same', color='c')
rects3 = ax.bar(x - width/2, A_bad, width, bottom=np.array(A_good) + np.array(A_same), label='Model A Bad', color='r')

# 绘制模型B的条形图
rects4 = ax.bar(x + width/2, B_good, width, label='Model B Good', color='g')
rects5 = ax.bar(x + width/2, B_same, width, bottom=B_good, label='Model B Same', color='y')
rects6 = ax.bar(x + width/2, B_bad, width, bottom=np.array(B_good) + np.array(B_same), label='Model B Bad', color='m')

# 添加一些文本标签
ax.set_xlabel('评价指标')
ax.set_ylabel('百分比')
ax.set_title('AB两个大模型在条理性，趣味性，性格一致性三方面的GSB对比')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()

plt.show()
```
