```python
# library
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create a dataset
df = pd.DataFrame(
    np.random.random((5, 5)),
    columns=["a", "b", "c", "d", "e"]
)

fig, axs = plt.subplots(
    nrows=2, ncols=2,
    figsize=(12, 10)
)

# plot using a color palette
sns.heatmap(df, cmap="coolwarm", ax=axs[0, 0])
sns.heatmap(df, cmap="coolwarm", ax=axs[0, 1])
sns.heatmap(df, cmap="coolwarm", ax=axs[1, 0])
sns.heatmap(df, cmap="coolwarm", ax=axs[1, 1])


fig.savefig(
    'heatmaps.svg', format='svg', dpi=300
)
plt.show()
```
