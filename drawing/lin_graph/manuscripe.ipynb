```python
import matplotlib.pyplot as plt

# 数据
labels = ['CommonCrawl', 'Book', 'Paper', 'Patent', 'Profession']
sizes = [30.5, 25.7, 21.5, 20.3, 1.81]
colors = ['#b0c4de', '#ffebcd', '#d3d3d3', '#dda0dd', '#add8e6']
explode = (0.1, 0, 0, 0, 0)  # 突出显示第一个分块

# 绘制饼图
fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
       shadow=False, startangle=90)
ax.axis('equal')  # 使饼图为正圆形

# 显示图像
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 数据
labels = ['CommonCrawl', 'Book', 'Paper', 'Patent', 'Profession']
sizes = [30.5, 25.7, 21.5, 20.3, 1.81]


# 定义不同的调色板
color_palettes = {
    "Set3": sns.color_palette("Set3"),
    "Pastel1": sns.color_palette("Pastel1"),
    "Dark2": sns.color_palette("Dark2"),
    "Accent": sns.color_palette("Accent"),
    "Paired": sns.color_palette("Paired")
}
colors = sns.color_palette("Pastel1")[:len(labels)]  # 使用Seaborn的Set3调色板

# 突出显示第一个分块
explode = (0.1, 0, 0, 0, 0)

# 绘制饼图
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(sizes, explode=explode, labels=labels, colors=colors, 
                                  autopct='%1.1f%%', shadow=False, startangle=90,
                                  textprops=dict(color="w"))

# 自定义字体和标签
for text in texts:
    text.set_fontsize(12)
    text.set_color('black')
for autotext in autotexts:
    autotext.set_fontsize(12)
    autotext.set_color('black')

ax.axis('equal')  # 使饼图为正圆形

# 显示图像
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.font_manager import FontProperties

# 数据
labels = ['CommonCrawl', 'Book', 'Paper', 'Patent', 'Profession']
sizes = [30.5, 25.7, 21.5, 20.3, 1.81]

# 定义不同的调色板
color_palettes = {
    "Set3": sns.color_palette("Set3")[:5],
    "Pastel1": sns.color_palette("Pastel1")[:5],
    "Dark2": sns.color_palette("Dark2")[:5],
    "Accent": sns.color_palette("Accent")[:5],
    "Paired": sns.color_palette("Paired")[:5]
}
colors = sns.color_palette("Pastel1")[:len(labels)]  # 使用Seaborn的Pastel1调色板

# 突出显示第一个分块
explode = (0.1, 0, 0, 0, 0)

# 自定义字体
# font_id = s
label_font = FontProperties(family='Bahnschrift', size=12, weight='bold')
# number_font = FontProperties(family='Consolas', size=12)

# 绘制饼图
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(sizes, explode=explode, labels=labels, colors=colors, 
                                  autopct='%1.1f%%', shadow=False, startangle=90,
                                  textprops=dict(color="black"))

# 自定义字体和标签
for text in texts:
    text.set_fontproperties(label_font)
    text.set_color('black')
for autotext in autotexts:
    # autotext.set_fontproperties(number_font)
    autotext.set_color('black')

# 调整标签位置
for text in texts:
    text.set_horizontalalignment('center')

ax.axis('equal')  # 使饼图为正圆形

# 设置标题
# plt.title("Distribution of Data Sources", fontsize=14, fontweight='bold', fontname='Ink Free')
plt.title("Distribution of Data Sources", fontsize=14, fontweight='bold', fontname='Microsoft New Tai Lue')

# 显示图像
plt.show()

```
